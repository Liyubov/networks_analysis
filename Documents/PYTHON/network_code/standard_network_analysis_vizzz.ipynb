```python

'''
Analyze networks
'''

print('drawing measures for graph')
plt.hist(list(nx.degree_centrality(Graph).values()))
plt.xlabel('normalized degree') #np.max(nx.degree_centrality(G_roads).values())*
plt.ylabel('frequency')
plt.show()

normdegrees = nx.degree_centrality(Graph).values()
#print(degrees)
print(len(set(normdegrees)))


plt.hist(list(nx.betweenness_centrality(Graph).values()))
plt.xlabel('betweenness')
plt.ylabel('frequency')
plt.show()



plt.hist(list(nx.edge_betweenness_centrality(Graph).values()))
plt.xlabel('edge betweenness')
plt.ylabel('frequency')
plt.show()



plt.hist(list(nx.clustering(Graph).values()))
plt.xlabel('clustering')
plt.ylabel('frequency')
plt.show()

print('average clustering')
avclus = nx.average_clustering(Graph)
print(avclus)

plt.hist(list(nx.closeness_centrality(Graph).values()))
plt.xlabel('closeness')
plt.ylabel('frequency')
plt.show()
```

```python
'''
Vizualize networks
'''

nx.draw(Graph)
plt.show()



print('plotting degree of nodes on a network')
deg_seq = nx.degree(Graph)
deg_size=[v * 10 for v in deg_seq]
nx.draw(Graph,  node_size = deg_size)
plt.show()
```
